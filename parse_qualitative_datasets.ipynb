```python
import torch
import dgl
from dgl.data.gindt import GINDataset

import networkx as nx
```

```python
dataset = GINDataset("REDDITBINARY", True)
```

```python
i = 0
for g, lbl in dataset:
    print(g, lbl)
    nxgraph = g.to_networkx()
    nx.write_gexf(nxgraph, 'data/REDDITBINARY/{}.{}.gexf'.format(i, lbl))
    i+= 1
```
